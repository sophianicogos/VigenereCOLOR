```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
!pwd
```

```python
## Function to preprocess UI image, then display it
def preprocess_and_disp(uiIMG_path):

    # Load uploaded image %%%have this uploaded elsewhere
    uiIMG = cv2.imread(uiIMG_path)

    plt.imshow(uiIMG)
    plt.axis('off')
    plt.title('Original Image')
    plt.show()

    # Stimulate spatial integration with Gaussian blur
    blurred_IMG = cv2.GaussianBlur(uiIMG, (7,7), 0)

    # Simulate contrast sensitivity with normalized contrast
    normalized_IMG = cv2.normalize(blurred_IMG, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)

    # Disp ui image and 'visually preprocessed' image
    #preprocessed_IMG = Image.fromarray(cv2.cvtColor(normalized_IMG, cv2.COLOR_BGR2RGB))

    # Display using Matplotlib
    plt.imshow(cv2.cvtColor(normalized_IMG, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title('Preprocessed Image')
    plt.show()

uiIMG_path = '/content/test.39.50.png'
preprocess_and_disp(uiIMG_path)

# # Load uiIMG for display
# uiIMG = cv2.imread(uiIMG_path)

# plt.imshow(uiIMG)
# plt.axis('off')
# plt.title('Original Image')
# plt.show()
```
